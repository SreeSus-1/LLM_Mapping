Code Execution:
To describe in a algorithmic format

Step-1: Import library called re (regular expression)

Step-2: Open the file from the payload folder (cnbc.html'), read content and store into "html_file_content"

Step-3: Now extract the content from the given html file using "fetch url" function.

fetch url function works as follows: To Extract the URLs from the given file, We used regular expression that identify the url which matches the following expression " http[s]?://(?:[a-zA-Z]|[0-9]|[-_@.&+]|[!*
$,]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'"

Step-4: Urls that match the reg ex format will be placed in urls variable and prints each url using for loop

Step-5: The total number of urls are counted and printed as output


Hence, anything that starts with "http" is printed in the output


```python
import re

def fetch_urls_from_html(file_path):
    with open(file_path, 'r') as file:
        html_file_content = file.read()
    url_match = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    urls = re.findall(url_match, html_file_content)
    return urls

file_path = 'cnbc.html'
urls = fetch_urls_from_html(file_path)

for url in urls:
    print(url)

url_count = len(urls)
print("Total Number Urls found in the given HTML file are:", url_count)
```
